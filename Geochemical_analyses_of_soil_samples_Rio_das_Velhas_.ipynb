# **2.0 - Localization of samples.**

We could see that the dataset has longitude and latitude information, for a better visualization of the data let's use a map to plot each soil sample. Then let's plot the limit area of Rio das Velhas Hydrographic basin  that is available here https://siga.cbhvelhas.org.br/siga-riodasvelhas/map#.


```python
# Load library

import folium

from folium.plugins import MarkerCluster
# The name of cloumns to get the correct name of latitude and longitude
soil_result.columns

# x is the Longitude  and y the Latitude, so let's use this to plot each point




# Create a base map of Brazil specifically of Rio das Velhas Location

rio_das_velhas_loc = [-18.802190, -44.205696]
map_soil_samples = folium.Map(location = rio_das_velhas_loc, zoom_start=7, tiles = "OpenStreetMap")
map_soil_samples
```

```python
# Let's Plot a all samples on a map

for i in range(0,len(soil_result)):
  folium.Marker([soil_result.iloc[i]["y"], soil_result.iloc[i]["x"]], popup=soil_result.iloc[i]["num_campo"]).add_to(map_soil_samples)


map_soil_samples
```

The first part is ok all samples were plotted in a folium map with an Open street map base map as background, and the start location was defined using the latitude and longitude of Rio das Velhas city. For the popup the field num_campo was chosen because it's a unique value for each sample, but now is necessary to plot the limit of the Hydrographic basin of Rio das Velhas, so let's read the shapefile with that limit, and the geopandas library will help to this task.   

```python
#Load Library

import geopandas as gpd

# reading the shapefile
rv_limit = gpd.read_file("/content/dataset_portifolio/cam_riodasvelhas_bacia.shp")
rv_limit.to_file("rv_limit.geojson", driver ="GeoJSON")

```

```python
# Open the GeoJson file

rv_limit_geojson = gpd.read_file("/content/dataset_portifolio/rv_limit.geojson")

rv_limit_geojson
```

```python
# now let's plot the limit area in folium map
folium.GeoJson(rv_limit_geojson).add_to(map_soil_samples)

map_soil_samples
```
